# Slicing :
in addition to accessing list elements one at a time , python provides a concise syntax to access the sublists . this is known as slicing process 


```python
nums = list(range(10))
print(nums)
print(nums[2:4])          # Get a slice from index 2 to 4 (exclusive); prints "[2, 3]"
print(nums[2:])           # Get a slice from index 2 to the end; prints "[2, 3, 4]"
print(nums[:2])           # Get a slice from the start to index 2 (exclusive); prints "[0, 1]"
print(nums[:])            # Get a slice of the whole list; prints "[0, 1, 2, 3, 4]"
print(nums[:-1])          # Slice indices can be negative; prints "[0, 1, 2, 3]"
nums[2:4] = [8, 9]        # Assign a new sublist to a slice
print(nums)               
```

# List Comprehension 
when programming , frequently , we want to transform one type of data in other type of data . As a example is listed below 

```python
nums = [0,1,2,3]
squares= []
for x in nums:
    squares.append(x**2)
print("Pre Comprehension")
print(nums)
print("Post Comprehension")
print(squares)
```

```python
# same stuff but one-liner

nums = [0,1,2,3]
squares= [x**2 for x in nums]

print("Pre Comprehension")
print(nums)
print("Post Comprehension")
print(squares)
```

```python
# list comprehension can also contain the containers 

nums = [0,23,343,5,66]
squares=[x**2 for x in nums if x % 2 == 0 ]
print("Numbers ")
print(nums)
print("Squared even numbers ")
print(squares)
```

### Dictionaries 
A dictionary stores (key, value) pairs, similar to a Map in Java or an object in Javascript. You can use it like this:




```python
felines = { 'tiger':'india ' , 'lion' : 'india ' , 'cheetah':'africa/india'}
print(felines)
print(felines.get('monkey', 'N/A'))  # Get an element with a default; prints "N/A"
print(felines.get('fish', 'N/A'))    # Get an element with a default; prints "wet"
```

# Sets 
A set is an unordered collection of distinct elements. As a simple example, consider the following:



```python
animals = {'cat', 'dog'}
print('cat' in animals)   # Check if an element is in a set; prints "True"
print('fish' in animals)  # prints "False"
animals.add('fish')       # Add an element to a set
print('fish' in animals)  # Prints "True"
print(len(animals))       # Number of elements in a set; prints "3"
animals.add('cat')        # Adding an element that is already in the set does nothing
print(len(animals))       # Prints "3"
animals.remove('cat')     # Remove an element from a set
print(len(animals))       # Prints "2"
```

# Tuples 
a tuple is an immutable ordered list of values . 
one of the most important differences is that tuples can be used as keys in dictionaries and as elements of sets, while lists cannot. Here is a trivial example:



```python
# create a dictionary with help of a tuple 

s = { (x, x+1) : x for x in range(10)}
t = (5,6)
print(type(t))
print(s[t])
print(s[(1,2)])
```

# Classes in Python :
The syntax for defining the classes in python is pretty straight forward 


---
In Python 3, this:
class Greeter(object):
and this:
class Greeter:
are effectively the same.
Why was object written?
In Python, every class inherits from another class.
object is the root/base class of almost all Python classes.


```python
class Animal(object):


    # constructor is the first method to be written in a class
    # is a special method that runs automatically when you create an object 

    def __init__(slf , name):
        slf.name = name
     #instance method 

    def meows(self , loud = False):
        if loud:
            print("Hello , %s ! Sorry for my Cat " % self.name.upper())
        else:
            print('Hello, %s' % self.name)

    
g = Animal("Fred")
g.meows()
g.meows(loud=True)
```
